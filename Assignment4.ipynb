```python
assignment["Score"].quantile ([0.25,0.75])
```

```python
q1,q3=assignment["Score"].quantile ([0.25,0.75])
print(q1,q3)
```

```python
iqr=q3-q1
```

```python
iqr
```

```python
lower_min=q1 - (1.5*iqr)
upper_max=q3+(1.5*iqr)
```

```python
print("Lower expected min of iqr=", lower_min)
print("Upper expected max of iqr=", upper_max)
```
