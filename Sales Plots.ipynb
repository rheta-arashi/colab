```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV dataset into a Pandas DataFrame
df = pd.read_csv('company_sales_data.csv')

# Calculate total profit for each month
total_profit = df.groupby('month_number')['total_units'].sum()

# Create the line plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(total_profit.index, total_profit.values, marker='o', color='red', linewidth=3)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Profit')
plt.title('Monthly Profits')

# Show the plot
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('company_sales_data.csv')

june_data = df[df['month_number'] == 6]

facecream_units = june_data['facecream'].values[0]
facewash_units = june_data['facewash'].values[0]

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(['Face Cream', 'Face Wash'], [facecream_units, facewash_units], color=['blue', 'blue'])
plt.xlabel('Product')
plt.ylabel('Number of Units Sold')
plt.title('Number of units Sold for Face Cream and Face Wash in June')
plt.show()
```
